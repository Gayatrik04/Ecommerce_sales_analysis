```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('../Dataset/Ecommerce_Sales_jan_june26.csv')
```

```python
df['Date']=pd.to_datetime(df['Date'])
df.info()
```

```python
total_orders = df['Order_ID'].count()
print(f'Total Orders: {total_orders}')

total_product_per_category = df.groupby('Category')['Product'].count()
print(f'\nTotal products per Cetegory',total_product_per_category)

Total_revenue=df['Total_price'].sum()
print(f'\n Total revenue:',Total_revenue)

Total_quantity_sold=df['Quantity'].sum()
print(f'\nTotal quantity sold:',Total_quantity_sold)

top_products=df.groupby('Product')['Total_price'].sum().sort_values(ascending=False).head(10)
print(f'\nTop 10 products:',top_products)

```

```python
category_quantity=df.groupby('Category')['Quantity'].sum().sort_values(ascending=True)
print(f'Quantity sold by each category:',category_quantity)

orders_by_category=df.groupby('Category')['Order_ID'].count().sort_values(ascending=True)
print(f'\nTotal Orders per category:',orders_by_category)
```

```python
city_by_revenue=df.groupby('City')['Total_price'].mean().sort_values(ascending=True)
plt.figure(figsize=(14,6))
plt.bar(city_by_revenue.index,city_by_revenue.values)
plt.title("Revenue by City's")
plt.xlabel('City')
plt.ylabel('Revenue')
plt.show()
```

```python
Revenue_by_Category=df.groupby('Category')['Total_price'].sum()
plt.pie(Revenue_by_Category.values,labels=Revenue_by_Category.index,autopct='%1.1f%%')
plt.title('Revenue By Category')
plt.show()
```

```python
product_revenue=df.groupby('Product')['Total_price'].sum().sort_values(ascending=False).head(5)

plt.bar(product_revenue.index,product_revenue.values)
plt.title('Top products with revenue')
plt.xlabel('Products')
plt.ylabel('Revenue')
plt.show()
```

```python
month_order = ['January','February','March','April','May','June']

monthly_revenue=df.groupby('Month')['Total_price'].sum().reindex(month_order)

plt.plot(monthly_revenue.index,monthly_revenue.values,marker='X',markerfacecolor='red',markeredgecolor='red')
plt.title('Monthly Sales')
plt.xlabel('Months')
plt.ylabel('Revenue')
plt.show()
```

```python
sns.heatmap(df[['Quantity','Price','Total_price']].corr(),annot=True,cmap='Blues')
plt.show()
```
